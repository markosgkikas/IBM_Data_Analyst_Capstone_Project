## Methodology:

The stack overflow survey provides the option to participants to select multiple Languages, Databases, Platforms, and Webframes that they have, or want to work with.

To be able to count these selections, we will need to explode "LanguageHaveWorkedWith", "DatabaseHaveWorkedWith", "PlatformHaveWorkedWith", "WebframeHaveWorkedWith", "LanguageWantToWorkWith", "DatabaseWantToWorkWith", "PlatformWantToWorkWith", "WebframeWantToWorkWith" into rows.

Here we are trying to count the frequency of each technology independently. We must not explode them in the same dataframe. Instead, we will create separate "mini-dataframes" for each category. This will keep row counts manageable and the data clean.

Otherwise, If a respondent has 5 languages, 5 databases, 5 platforms, and 5 webframes, that one person will turn into $5 \times 5 \times 5 \times 5 = 625$ rows. This will affect running time exponentially.

The Big O notation for the number of rows generated would be: $$O(N \cdot M^K)$$

Where:

$N$ = The original number of rows in the dataframe,

$M$ = The average number of items in each list,

$K$ = The number of columns that are being exploded.

## Methodology:

For practical and dashboard aesthetic reasons, we will trim down the value labels of Age, EdLevel and Country.

E.g. "35-44 years old" --> "35-44", "Bachelor's degree (B.A., B.S., B.Eng., etc.)" --> "Bachelor" and "United States of America" --> "USA".

### Step 1: Create maps or replace value tags

```python
age_edlevel_country_df['Age'].unique()
```

```python
age_map = {
    '35-44 years old' : '35-44' ,
    '45-54 years old' : '45-54' ,
    '25-34 years old' : '25-34' ,
    '55-64 years old' : '55-64' ,
    '18-24 years old' : '18-24' ,
    'Under 18 years old' : '<18' ,
    '65 years or older' : '>=65' ,
    'Prefer not to say' : 'Prefer not to say'
}
```

```python
age_edlevel_country_df['EdLevel'].unique()
```

```python
edlevel_map = {
    "Bachelor's degree (B.A., B.S., B.Eng., etc.)" : 'Bachelor' ,
    "Master's degree (M.A., M.S., M.Eng., MBA, etc.)" : 'Master' ,
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)' : 'Professional degree' ,
    'Some college/university study without earning a degree' : 'College/University Student' ,
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)' : 'Secondary school' ,
    'Associate degree (A.A., A.S., etc.)' : 'Associate degree' ,
    'Something else' : 'Something else' ,
    'Primary/elementary school' : 'Primary/Elementary School' ,
}
```

```python
age_edlevel_country_df['Country'].unique()
```

### Step 2: Replace value labels

```python
age_edlevel_country_df['Age'] = age_edlevel_country_df['Age'].map(age_map)
```

```python
age_edlevel_country_df['EdLevel'] = age_edlevel_country_df['EdLevel'].map(edlevel_map)
```

```python
age_edlevel_country_df['Country'] = age_edlevel_country_df['Country'].replace('United Kingdom of Great Britain and Northern Ireland','UK')
age_edlevel_country_df['Country'] = age_edlevel_country_df['Country'].replace('United States of America','USA')
```
